```
import pandas as pd
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM,Bidirectional
from tensorflow.keras.layers import Dense,Dropout
```

```
df=pd.read_csv('/content/drive/My Drive/LSTM Kaggle fake news dataset/train.csv',engine='python', error_bad_lines=False)
```

```
df.head()
```

```
# Check Nan Value in title colum
df['title'].isnull().sum()
```

```
# Check Nan Value in whole train dataset
df.isnull().sum()
```

```
###Drop Nan Values
df=df.dropna()
```

```
# Deleted Nan/Null values from dataset
df.isnull().sum()
```

```
## Get the Independent Features

X=df.drop('label',axis=1)
```

```
## Get the Dependent features
y=df['label']
```

```
X.shape
```

```
y.shape
```

```
### fix the Vocabulary size
voc_size=5000
```

```
messages=X.copy()
```

```
messages.head()
```

```
messages['title'][1]
```

```
messages.reset_index(inplace=True)
```

```
messages.head()
```

```
# Data preprocessing
import nltk
import re
from nltk.corpus import stopwords
```

```
nltk.download('stopwords')
```

```
#dataset preprocesssing
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i]) # remove all other than alphabet & add spece
    review = review.lower()    # convert the dataset in lower case
    review = review.split()    # split the word by space  
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```
corpus
```

```
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

```
# Embedding Representation -paddin for making same length of sentance
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```
embedded_docs[0]
```

```
embedded_docs.astype
```

```
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Bidirectional(LSTM(100)))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```
len(embedded_docs),y.shape
```

```
# converting to numpy array
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```
X_final.astype
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_final,y_final,test_size=0.33,random_state=42)
```

```
# model Traning
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=60)
```

```
# Performance Metrics And Accuracy¶
y_pred=model.predict_classes(X_test)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
